```python
import numpy as np
from time import time
from driver import Model
```

```python
myip = Model("bitstream/cnn.bit")
```

```python
import pandas as pd
data = pd.read_csv('Dataset/grenade24.csv', sep=',', header=None)
data = np.int16(data).reshape(-1, 6)
print(data)
gold_result = 2
```

```python
# sliding window with window size 15, stride 5
data_length = len(data)
window_size = 15
window_stride = 5

highest_false_positive_confidence = 0

for j in range(0, data_length - window_size + 1, window_stride):
    predicted, softmax_confidence = myip.inference(data[j:j + window_size].flatten())
#     print(f"raw outputs: {myip.raw_outputs}")
    print(f"prediction: {predicted} vs GOLD: {gold_result} with confidence {softmax_confidence * 100}%")
    print(f"remaining number of frames required to fill the CNN window: {int(myip.register_map.data_required)}")
    print("\n")

    if predicted != gold_result and softmax_confidence > highest_false_positive_confidence:
        highest_false_positive_confidence = softmax_confidence
myip.resetBuffer()

# if the confidence goes higher than this value then we do not need any more input
print(f"best threshold = {highest_false_positive_confidence*100}%")
```
